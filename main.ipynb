```python
from pathlib import Path

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
from tqdm import tqdm

import snntorch as snn
from snntorch import surrogate

from sklearn.metrics import confusion_matrix, classification_report
```

## Hyperparameters

```python
H = {
    'batch_size': 128,
    'lr': 1e-3,
    'epochs': 10, 
    'device': 'cuda' if torch.cuda.is_available() else 'cpu',
    'snn_timesteps': 25,
}


SAVE_DIR = Path("runs")
SAVE_DIR.mkdir(exist_ok=True)
```

## Data

```python
transform = transforms.Compose([transforms.ToTensor()])


def get_train_loader(batch_size=128):
    train_set = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
    train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=True, num_workers=2)
    return train_loader

def get_test_loader(batch_size=128):
    test_set = datasets.MNIST(root='./data', train=False, download=True, transform=transform)
    test_loader = DataLoader(test_set, batch_size=batch_size, shuffle=False)
    return test_loader
```

## CNN 

```python
class CNN(nn.Module):
    def __init__(self, num_classes=10):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(1, 12, kernel_size=5, padding=2),
            nn.ReL  U(),
            nn.MaxPool2d(2),
            nn.Conv2d(12, 32, kernel_size=5, padding=2),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(32 * 7 * 7, 256),
            nn.ReLU(),
            nn.Linear(256, num_classes),
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x
```

## SNN

```python
class ConvSNN(nn.Module):
    def __init__(self, time_steps=20):
        super().__init__()
        self.T = time_steps
        self.conv1 = nn.Conv2d(1, 12, 5, padding=2)
        self.conv2 = nn.Conv2d(12, 32, 5, padding=2)
        self.pool = nn.MaxPool2d(2)
        self.pool2 = nn.MaxPool2d(2)
        self.fc1 = nn.Linear(32 * 7 * 7, 256)
        self.fc2 = nn.Linear(256, 10)

        self.lif1 = snn.Leaky(beta=0.9, spike_grad=surrogate.fast_sigmoid())
        self.lif2 = snn.Leaky(beta=0.9, spike_grad=surrogate.fast_sigmoid())
        self.lif3 = snn.Leaky(beta=0.9, spike_grad=surrogate.fast_sigmoid())

    def forward(self, x, record_spikes=False):
        mem1 = self.lif1.init_leaky()
        mem2 = self.lif2.init_leaky()
        mem3 = self.lif3.init_leaky()

        spk_sum = torch.zeros(x.size(0), 10, device=x.device)

        if record_spikes:
            total_spikes = {'conv1': 0, 'conv2': 0, 'fc1': 0, 'fc2': 0}

        for _ in range(self.T):
            out = torch.relu(self.conv1(x))
            out = self.pool(out)
            spk1, mem1 = self.lif1(out, mem1)

            out2 = torch.relu(self.conv2(spk1))
            out2 = self.pool2(out2)
            spk2, mem2 = self.lif2(out2, mem2)

            h = self.fc1(spk2.view(spk2.size(0), -1))
            spk3, mem3 = self.lif3(h, mem3)
            logits = self.fc2(spk3)
            spk_sum += logits

            if record_spikes:
                total_spikes['conv1'] += spk1.sum().item()
                total_spikes['conv2'] += spk2.sum().item()
                total_spikes['fc1'] += spk3.sum().item()

        if record_spikes:
            for k in total_spikes:
                total_spikes[k] /= (x.size(0) * self.T)
            return spk_sum / self.T, total_spikes
        else:
            return spk_sum / self.T
```

## Train/Eval

```python
def evaluate(model, dataloader, device):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for imgs, labels in dataloader:
            imgs, labels = imgs.to(device), labels.to(device)
            logits = model(imgs)
            preds = logits.argmax(dim=1)
            correct += (preds == labels).sum().item()
            total += labels.numel()
    return correct / total
```

```python
def train_cnn(model, train_loader, test_loader, device, epochs=5, lr=1e-3, save_name="cnn.pth"):
    model = model.to(device)
    opt = optim.Adam(model.parameters(), lr=lr)
    criterion = nn.CrossEntropyLoss()


    for ep in range(epochs):
        model.train()
        pbar = tqdm(train_loader, desc=f"Train Epoch {ep}")
        running_loss = 0.0
        for imgs, labels in pbar:
            imgs, labels = imgs.to(device), labels.to(device)
            opt.zero_grad()
            logits = model(imgs)
            loss = criterion(logits, labels)
            loss.backward()
            opt.step()
            running_loss += loss.item()
            pbar.set_postfix(loss=running_loss)


        acc = evaluate(model, test_loader, device)
        print(f"Epoch {ep} test acc: {acc:.4f}")
    
    torch.save(model.state_dict(), SAVE_DIR / save_name)
    return model
```

```python
def train_snn(model, train_loader, test_loader, device, epochs=5, lr=1e-3, save_name="snn.pth"):
    model = model.to(device)
    opt = optim.Adam(model.parameters(), lr=lr)
    criterion = nn.CrossEntropyLoss()


    for ep in range(epochs):
        model.train()
        pbar = tqdm(train_loader, desc=f"SNN Train Epoch {ep}")
        running_loss = 0.0
        for imgs, labels in pbar:
            imgs, labels = imgs.to(device), labels.to(device)
            opt.zero_grad()
            out = model(imgs)
            loss = criterion(out, labels)
            loss.backward()
            opt.step()
            running_loss += loss.item()
            pbar.set_postfix(loss=running_loss)
    
        acc = evaluate(model, test_loader, device)
        print(f"Epoch {ep} test acc: {acc:.4f}")

    torch.save(model.state_dict(), SAVE_DIR / save_name)
    return model

```

## Plotting

```python
def plot_sample_raster(model, test_loader, device, num_samples=3):
    model.eval()
    imgs, labels = next(iter(test_loader))
    imgs = imgs[:num_samples].to(device)
    labels = labels[:num_samples]

    if hasattr(model, 'T'):
        T = model.T
        batch = imgs.size(0)
        out_acc = torch.zeros(batch, 10, T)
        if isinstance(model, globals().get('ConvSNN')):
            mem1 = model.lif1.init_leaky()
            mem2 = model.lif2.init_leaky()
            mem3 = model.lif3.init_leaky()
            for t in range(T):
                with torch.no_grad():
                    out = model.conv1(imgs)
                    out = torch.relu(out)
                    out = model.pool(out)
                    spk1, mem1 = model.lif1(out.view(out.size(0), -1), mem1)
                    out2 = model.conv2(out)
                    out2 = torch.relu(out2)
                    out2 = model.pool2(out2)
                    spk2, mem2 = model.lif2(out2.view(out2.size(0), -1), mem2)
                    h = model.fc1(spk2)
                    spk3, mem3 = model.lif3(h, mem3)
                    logits = model.fc2(spk3)
                    out_acc[:, :, t] = logits.cpu()
        else:
            for t in range(T):
                with torch.no_grad():
                    logits = model(imgs)
                    out_acc[:, :, t] = logits.cpu()


        # plot
        for i in range(num_samples):
            plt.figure(figsize=(8,3))
            plt.subplot(1,2,1)
            plt.imshow(imgs[i,0].cpu(), cmap='gray')
            plt.title(f"Label: {labels[i].item()}")
            plt.axis('off')


            plt.subplot(1,2,2)
            plt.imshow(out_acc[i].T, aspect='auto')
            plt.xlabel('class index')
            plt.ylabel('timesteps')
            plt.title('Logit over timesteps')
            plt.colorbar()
            plt.tight_layout()
            plt.show()

    else:
        with torch.no_grad():
            logits = model(imgs)
            probs = torch.softmax(logits, dim=1)
        for i in range(num_samples):
            plt.figure(figsize=(6,3))
            plt.subplot(1,2,1)
            plt.imshow(imgs[i,0].cpu(), cmap='gray')
            plt.title(f"Label: {labels[i].item()}")
            plt.axis('off')
            plt.subplot(1,2,2)
            plt.bar(range(10), probs[i].cpu())
            plt.title('Predicted probs')
            plt.tight_layout()
            plt.show()
```

## Metrics

```python
def evaluate_metrics(model, loader, device, record_spikes=False):
    model.eval()
    all_labels, all_preds = [], []
    spike_stats = {'conv1': 0, 'conv2': 0, 'fc1': 0, 'fc2': 0}
    batches = 0

    with torch.no_grad():
        for imgs, labels in tqdm(loader, desc='Evaluating'):
            imgs, labels = imgs.to(device), labels.to(device)

            if record_spikes:
                logits, spikes = model(imgs, record_spikes=True)
                for k in spike_stats:
                    spike_stats[k] += spikes[k]
                batches += 1 
            else:
                logits = model(imgs)

            preds = logits.argmax(dim=1)
            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())

    cm = confusion_matrix(all_labels, all_preds)
    report = classification_report(all_labels, all_preds, digits=4)
    acc = np.trace(cm) / np.sum(cm)

    if record_spikes and batches > 0:
        for k in spike_stats:
            spike_stats[k] /= batches
        total_spikes = sum(spike_stats.values())
        energy_proxy = total_spikes
    else:
        spike_stats = None
        energy_proxy = None

    return acc, cm, report, spike_stats, energy_proxy
```

```python
def plot_cm(cm):
    plt.figure(figsize=(6, 5))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title("Confusion Matrix")
    plt.colorbar()
    plt.xticks(range(len(set(cm[0]))), set(cm[0]))
    plt.yticks(range(len(set(cm[0]))), set(cm[0]))
    plt.xlabel('Predicted label')
    plt.ylabel('True label')

    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, cm[i, j], ha='center', va='center', color='red')

    plt.show()
```

## Energy Comparison

```python
def compare_energy(cnn_model, snn_model, loader, device):
    print("\nEvaluating CNN...")
    acc_cnn, cm_cnn, report_cnn, _, _ = evaluate_metrics(cnn_model, loader, device)
    print(f"CNN Accuracy: {acc_cnn*100:.2f}%\n")

    print("Evaluating SNN...")
    acc_snn, cm_snn, report_snn, spikes, energy_snn = evaluate_metrics(
        snn_model, loader, device, record_spikes=True
    )
    print(f"SNN Accuracy: {acc_snn*100:.2f}%\n")

    if spikes is not None:
        total_samples = len(loader.dataset)
        total_spikes = sum(spikes[k] for k in spikes)
        energy_snn = total_spikes / total_samples 

    cnn_energy = 1.0
    snn_energy_norm = energy_snn / cnn_energy if energy_snn is not None else None

    print("\n--- ENERGY EFFICIENCY COMPARISON ---")
    print(f"CNN baseline energy (normalized): {cnn_energy:.2f}")
    print(f"SNN normalized spike-based energy proxy: {energy_snn:.2f}")
    print(f"Relative efficiency (SNN/CNN): {snn_energy_norm:.2f}")

    # Plot bar chart
    plt.figure(figsize=(6, 4))
    plt.bar(['CNN (baseline)', 'SNN (proxy)'], [cnn_energy, energy_snn], color=['gray', 'orange'])
    plt.ylabel('Normalized Energy Proxy')
    plt.title('CNN vs SNN Energy Efficiency Comparison')
    plt.show()
```

## Main

```python
device = H['device']
train_loader = get_train_loader(H['batch_size'])
test_loader = get_test_loader(H['batch_size'])
```

```python
cnn_model = CNN()
print('Training CNN baseline on device:', device)
cnn_model = train_cnn(cnn_model, train_loader, test_loader, device, epochs=H['epochs'], lr=H['lr'], save_name='cnn_baseline.pth')
acc = evaluate(cnn_model, test_loader, device)
print('CNN test accuracy:', acc)
plot_sample_raster(cnn_model, test_loader, device)
```

```python
snn_model = ConvSNN(time_steps=H['snn_timesteps'])
snn_model = train_snn(snn_model, train_loader, test_loader, device, epochs=H['epochs'], lr=H['lr'], save_name='snn_conv.pth')
acc = evaluate(snn_model, test_loader, device)
print('SNN test accuracy (approx):', acc)
plot_sample_raster(snn_model, test_loader, device)
```
